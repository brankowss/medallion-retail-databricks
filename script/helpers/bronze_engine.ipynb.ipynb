# Bronze Ingestion Engine
Standardized logic for CSV to Delta Lake ingestion with ingestion timestamp.

```python
%python
from pyspark.sql.functions import current_timestamp

def ingest_bronze(config_list):
    """
    Ingests multiple CSV files into Delta tables based on a config list.
    """
    for item in config_list:
        print(f"Reading from: {item['path']}")
        
        df = (
            spark.read
            .option("header", True)
            .option("inferSchema", True)
            .csv(item["path"])
            .withColumn("ingestion_ts", current_timestamp())
        )

        target_table = f"workspace.bronze.{item['table']}"
        df.write.format("delta").mode("overwrite").saveAsTable(target_table)
        print(f"Successfully saved to: {target_table}")
```
